```r
attach(Boston)
lm.fit=lm(medv~lstat)
lm.fit
```

```r
summary(lm.fit)
```

```r
names(lm.fit)
```

```r
coef(lm.fit)
```

```r
# confidence interval
confint(lm.fit)
```

```r
conf <- predict(lm.fit,data.frame(lstat=Boston$lstat), interval="confidence")
pred <- predict(lm.fit,data.frame(lstat=Boston$lstat), interval="prediction") 
lwr = pred[,2]
upr = pred[,3]
```

```r
library("ggplot2")
p <- ggplot(Boston, aes(lstat, medv)) +
  geom_point() +
  stat_smooth(method = lm)
```

```r
# 3. Add prediction intervals
p + geom_line(aes(y = lwr), color = "red", linetype = "dashed")+
    geom_line(aes(y = upr), color = "red", linetype = "dashed")
```

```r
# basic plot function in R
plot(lstat,medv)
abline(lm.fit)
abline(lm.fit,lwd=3)
abline(lm.fit,lwd=3,col="red")
plot(lstat,medv,col="red")
plot(lstat,medv,pch=20)
plot(lstat,medv,pch="+")
plot(1:20,1:20,pch=1:20)
```

```r
par(mfrow=c(2,2))
plot(lm.fit)
```

```r
par(mfrow=c(2,2))
plot(predict(lm.fit), residuals(lm.fit))
plot(predict(lm.fit), rstudent(lm.fit)) # studentized residual https://en.wikipedia.org/wiki/Studentized_residual
plot(hatvalues(lm.fit)) # leverage statistics h_i
which.max(hatvalues(lm.fit))
```

```r
# Multiple Linear Regression

lm.fit=lm(medv~lstat+age,data=Boston)
summary(lm.fit)
```

```r
lm.fit=lm(medv~.,data=Boston)
summary(lm.fit)
```

```r
library(car)
vif(lm.fit) # variance inflation factor (detect multi-colinearity)
```

```r
lm.fit1=lm(medv~.-age,data=Boston) #all predictors except age
summary(lm.fit1)
```

```r
lm.fit1=update(lm.fit, ~.-age)
summary(lm.fit1)
```

```r
# Interaction Terms

summary(lm(medv~lstat*age,data=Boston))
```

```r
# Non-linear Transformations of the Predictors

lm.fit2=lm(medv~lstat+I(lstat^2))
summary(lm.fit2)
```

```r
lm.fit=lm(medv~lstat,data=Boston)
anova(lm.fit,lm.fit2) # indicate non-linearity
```

```r
par(mfrow=c(2,2))
plot(lm.fit2) #cook's distance https://en.wikipedia.org/wiki/Cook%27s_distance
```

```r
lm.fit5=lm(medv~poly(lstat,5))
summary(lm.fit5)
```

```r
summary(lm(medv~log(rm),data=Boston))
```
